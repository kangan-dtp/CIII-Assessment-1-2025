# Car Batteries Tracker #

You're starting to manage your car batteries inventory with a new console-based Python application: **Car Batteries Tracker**.

---

## Features ##
1. **Add batteries to the inventory**  
   - Record the quantity of batteries added along with the date of addition.
2. **Install batteries**  
   - Record the quantity of batteries installed, the car registration, and the date of installation.
3. **View current inventory levels**  
   - Display the total number of batteries currently in stock.
4. **View usage history**  
   - Display a history of all batteries added and installed, including dates and car registrations.
5. **Exit the program**  
   - Safely exit the application.

---

## Rules ##
The inventory starts with 50 batteries and the following usages:

     - added: 30 batteries on 01-01-2025
     - installed: 3 batteries for registration KBT-678 on 01-02-2025
     - installed: 2 batteries for registration HDT-903 on 05-05-2025


- When batteries are added, the program records the quantity and the date of addition.  
- When batteries are installed, the program checks if there is enough quantity in stock. If not, it displays a warning message and does not allow the transaction. When installing batteries, the program records the quantity installed, the car registration, and the date of installation.  
- Usage history includes all additions and installations, along with their respective dates and car registrations.
